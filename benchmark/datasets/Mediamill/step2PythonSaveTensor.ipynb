```python
import torch
A = torch.tensor(df.iloc[:, 0: 120].to_numpy())
B = torch.tensor(df.iloc[:, 120: 221].to_numpy())
```

```python
# save to tensor that C++ can load directly

from torch import nn
class TensorContainer(nn.Module):
    def __init__(self, tensor_dict):
        super().__init__()
        for key,value in tensor_dict.items():
            setattr(self, key, value)

tensor_dict = {
    "A": A,
    "B": B
}
tensors = TensorContainer(tensor_dict)
tensors = torch.jit.script(tensors)
tensors.save('mediamill.pth')
```
